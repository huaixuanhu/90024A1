## Current local counting rules

- `Mastodon` uses `doc.language`
- `BlueSky` uses `record.langs`
- `str` counts as one language and `list` counts each language item
- `None`, `null`, empty strings, and empty lists are skipped and tracked
- All language codes are normalized to lowercase
- Suspicious non-standard codes are preserved in raw form and reported separately
